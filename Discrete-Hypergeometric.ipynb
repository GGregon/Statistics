**Digressão quanto as notações**
* $X$ é a variável aleatória (função que associa subconjuntos de $S$ nos $\mathbb{R}$ - $X:S\rightarrow \mathbb{R}$
* $\mathcal{X}$ é a **imagem** da variável aleatória $X$. Isto é, o conjunto de todos os valores de $X(s_i)$ pode assumir
* $P$ ou também $P(s)$ é a função de probabilidade definida sobre o espaço amostral
* $P_X$ ou também $P_X(X=x)$ é a função de probabilidade definida sobre a variável aleatória $X$. O subscrito é para enfatizar que se trata de uma probabilidade induzida, pois seu cálculo é definido indiretamente.

## 1. Definição do Espaço Amostral $S$
Seja um experimento aleatório cujo espaço amostral contém $N$ elementos, dos quais $M$ compartilham de uma mesma característica $\mathcal{C}_a $ e os $N-M$ restantes compartilham de, digamos, $\mathcal{C}_b$. Por exemplo, $M$ elementos são azuis e $N$ são verdes. O espaço amostral desse experimento será definido como o conjunto de todas as possíveis amostras de tamanho $k$ retiradas da população.

$$
S = \{s_1, s_2, ..., s_m\}
$$

onde cada $ s_i $ representa uma amostra de tamanho $k$ retirada da população de $ N $ elementos. O número total de elementos no espaço amostral é:

$$
m = \binom{N}{k} = \frac{N!}{k!(N-k)!}
$$

Esse valor representa **todas as combinações possíveis** de $k$ elementos que podemos retirar de $ N $.

> A **distribuição hipergeométrica** modela um experimento onde temos uma população finita $ N $, contendo $ M $ sucessos e $ N - M $ fracassos. Extraímos uma amostra de tamanho $ k $, **sem reposição**, e contamos o número de sucessos $ x $ na amostra.

## 2. Definição da Função de Probabilidade $ P $ sobre $ S $
Para definir $P$ precisamos assumir algumas suposições. No caso do experimento descrito acima, a suposição é que todas as amostras de tamanho $k$ são igualmente prováveis de serem selecionadas. Matematicamente, isso se traduz em $P({S=s_i}) = P({S=s_j}) \quad \forall i,j \in \{1,\cdots,m\}$. Sendo $P$ uma função de probabilidade, ela deve satisfazer os axiomas de *Koglomorov*, mas não iremos mostrar a validade de todos eles, mas um deles exige que a soma das probabilidades de eventos disjuntos seja igual a 1. Nesse caso, usando a premissa da distribuição uniforme, concluimos que:

$$
P(\{S=s_i\}) = \frac{1}{\binom{N}{k}}, \quad \forall s_i \in S
$$

onde $ \binom{N}{k} $ é o **número total de amostras possíveis**, ou seja, todas as formas de escolher $ k $ elementos de $ N $.

## 3. Definição da Variável Aleatória $ X $
Agora que temos o espaço amostral $S$ e a **função de probabilidade $ P $ definida sobre $ S $**, podemos definir a variável aleatória $ X $.

A variável aleatória $ X $ será definida como o número de observações da classe $\mathcal{C}_a$ (sem perda de generalidade) na amostra de tamanho $ k $**. Assim, $ X $ é uma função que mapeia cada amostra $ s_i $ do espaço amostral $ S $ para um número real:

$$
X: S \to \mathbb{R}
$$

onde,

$$
X(s_i) = x
$$

significa que a amostra $ s_i $ contém exatamente $x$ elementos da classe $\mathcal{C}_a$. Em outras palavras, $ X(s_i) $ conta **quantos dos $ k $ elementos sorteados pertencem a classe** $\mathcal{C}_a$.

## 4. Probabilidade Induzida $ P_X(X = x) $
Agora que definimos $ X $, podemos calcular a **distribuição de proababilidade** de $X$, $ P_X(X = x) $, que é dada por

$$
P_X(X = x) = P(\{s_i \in S: X(s_i) = x\})
$$
Lembrando que essa equação significa:  
> _"A probabilidade de $ X $ assumir um determinado valor $ x $ é igual à soma das probabilidades dos elementos de $ S $ que são mapeados para $ x $ pela função $ X $."_

Cada amostra válida contém **$ x $ elementos da classe $\mathcal{C}_a$ extraídos dos $ M $ disponíveis** e **$ k - x $ elementos da classe complementar $\mathcal{C}_b$ extraídos dos $ N - M $** disponíveis. O número total dessas amostras é dado pelo produto de coeficientes binomiais:

$$
\# \text{amostras com } x \text{ elementos da classe } \mathcal{C}_a = \binom{M}{x} \binom{N - M}{k - x}
$$

Como todas as amostras em $ S $ têm a **mesma probabilidade $ P(s_i) = \frac{1}{\binom{N}{k}} $**, a **probabilidade de $ X = x $** é:

$$
P_X(X = x) = \frac{\binom{M}{x} \binom{N - M}{k - x}}{\binom{N}{k}}, \quad \max(0, k - (N - M)) \leq x \leq \min(k, M)
$$

onde:
- $ \binom{M}{x} $ representa o número de maneiras de escolher $ x $ sucessos da população de $ M $ sucessos.
- $ \binom{N - M}{k - x} $ representa o número de maneiras de escolher $ k - x $ fracassos da população de $ N - M $.
- $ \binom{N}{k} $ é o número total de amostras possíveis de tamanho $ k $.

## 5. Probabilidade acumulada $P_X(X\le x)$
A CDF, $F_X(x)$ é definida como:

$$
F(x) = P(X \leq x) = \sum_{j=0}^{x} P_X(X = j)
$$

Ou seja, $F_X(x)$ é a soma das probabilidades da PMF até o valor $ x $. Para valores fora do intervalo válido de $ X $:

$$
F(x) =
\begin{cases}
    0, & x < 0 \\
    \sum_{j=0}^{\lfloor x \rfloor} P_X(X = j), & 0 \leq x \leq \min(k, M) \\
    1, & x > \min(k, M)
\end{cases}
$$

Isso garante que a **CDF** cresce de forma discreta, atingindo 1 no maior valor possível de $ X $.


## 6. Momentos da distribuição
Momentos são estatísticas úteis para se caracterizar a distribuição de uma variável aleatória e apesar de, em alguns casos, ser possível calcular infinitos deles, nos são uteis somente os dois primeiros: a média e a variância.

### 6.1. Média ($\mu$) - Primeiro momento ($\mathbb{E}X^1 = \mu_1'$)

$$
\mathbb{E}X = \sum_{x \in X(S)} x P_X(X = x)
$$

Para a **distribuição hipergeométrica** com parâmetros $N,M$ e $k$, temos:
$$
\mathbb{E}X = \frac{M}{N}k
$$

### 6.2. Variância $\text{Var}(X)$ - Segundo momento central ($\mathbb{E}(X-\mu)^2 = \mu_2)$

Usando um resultado (simples de provar usando as propriedades do valor esperado), temos:
$$
\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

A expressão final (cuja dedução foge do objetivo deste notebook) fica:
$$
\text{Var}(X) = k\frac{M}{N} \frac{(N-M)(N-k)}{N(N-1)}
$$


## Visualizações
Para visualizar os gráficos abaixo, irei contextualizar com um exemplo prático onde essa distribição é utilizada: controle de qualidade (amostragem de aceitação). Considere um lote com $N$ maçãs, das quais $M$ estão podres (classe $\mathcal{C}_a$) e, consequentemente, $N-M$ estão boas (classe $\mathcal{C}_b$). Vamos analisar a probabilidade de uma amostra aleatória contendo $k$ maçãs conter exatamente $x$ maçãs podres.

```python
# Importando bibliotecas e classes uteis
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Parâmetros da distribuição hipergeométrica
N, M, k = 50, 12, 20

# Criando valores para X (número de sucessos na amostra)
x_values = np.arange(max(0, k - (N - M)), min(k, M) + 1)

# Calculando PMF e CDF
pmf_values = stats.hypergeom.pmf(x_values, N, M, k)
cdf_values = stats.hypergeom.cdf(x_values, N, M, k)

# Cálculo dos momentos
mean_x = k * (M / N)
var_x = k * (M / N) * ((N - M) / N) * ((N - k) / (N - 1))
std_x = np.sqrt(var_x)  # Desvio padrão
```

Para o problema modelado, consideramos:
* $N$ = 50 maçãs no total
* $M$ = 12 maçãs podres
* $k$ = 20 maçãs selecionadas aleatoriamente

```python
# Criando os gráficos
f, (ax_pmf, ax_cdf) = plt.subplots(1, 2, figsize=(13, 4), layout='constrained')

# Função de massa de probabilidade (PMF)
ax_pmf.stem(x_values, pmf_values, basefmt="C0-")
ax_pmf.set_title(r"$P_X(X=x)$")
ax_pmf.errorbar(mean_x, max(pmf_values) / 2, xerr=std_x, fmt='o', color='tab:blue', capsize=5, alpha=0.5)

# Função de probabilidade acumulada (CDF)
ax_cdf.step(x_values, cdf_values, where="mid", label="CDF", linewidth=2)
ax_cdf.set_title(r"$F_X(X)$")

# Ajustes gerais nos gráficos
for ax in (ax_pmf, ax_cdf):
    ax.grid(alpha=0.5)
    ax.set_xlabel('X = Número de maçãs podres')
    ax.set_xticks(x_values)

plt.show()
```

Interpretação:
* A probabilidade de haver 5 maçãs podres é de ~26%
* A probabilidade de todas as maçãs estarem podres é 0.001%
* A probabilidade de nenhuma maçã estar podre é 0.07%

Por fim, irei simular o efeito de cada um dos parâmetros da distribuição. Para isso, irei fixar dois deles e variar o outro, com o objetivo de ilustrar o impacto de cada um deles. Contudo, podemos fazer algumas "previsões" a partir das expressões dos momentos da distribuição hipergeométrica.

$$
\mathbb{E}X = \frac{KM}{N}
$$

$$
\text{Var}(X) = k\frac{M}{N} \frac{(N-M)(N-k)}{N(N-1)}
$$

```python
# Definindo parâmetros base
N_base, M_base, k_base = 50, 12, 20

# Escolhendo novos valores para cada parâmetro enquanto mantém os outros fixos
k_values = [10, 20, 30]  # Variando k, mantendo N e M fixos
M_values = [8, 12, 16]   # Variando M, mantendo N e k fixos
N_values = [40, 50, 60]  # Variando N, mantendo M e k fixos

# Definindo cores diferentes para as curvas
cores = ['tab:blue', 'tab:red', 'black']
```

```python
# Criando a figura com três subgráficos
fig, axes = plt.subplots(1, 3, figsize=(18, 8), layout='constrained')

# Variando k, mantendo N e M fixos
for idx, k in enumerate(k_values):
    x_values = np.arange(max(0, k - (N_base - M_base)), min(k, M_base) + 1)
    pmf_values = stats.hypergeom.pmf(x_values, N_base, M_base, k)
    stem_container = axes[0].stem(x_values, pmf_values, basefmt="C0-", markerfmt='o', linefmt='-')
    plt.setp(stem_container.markerline, color=cores[idx])
    plt.setp(stem_container.stemlines, color=cores[idx])
    axes[0].plot(x_values, pmf_values, color=cores[idx], label=f'k={k}')

axes[0].set_title(r"Variação de $k$ (Mantendo $N, M$ fixos)")
axes[0].set_xlabel('X')
axes[0].set_ylabel(r"$P_X(X=x)$")
axes[0].legend()
axes[0].grid(alpha=0.5)

# Variando M, mantendo N e k fixos
for idx, M in enumerate(M_values):
    x_values = np.arange(max(0, k_base - (N_base - M)), min(k_base, M) + 1)
    pmf_values = stats.hypergeom.pmf(x_values, N_base, M, k_base)
    stem_container = axes[1].stem(x_values, pmf_values, basefmt="C0-", markerfmt='o', linefmt='-')
    plt.setp(stem_container.markerline, color=cores[idx])
    plt.setp(stem_container.stemlines, color=cores[idx])
    axes[1].plot(x_values, pmf_values, color=cores[idx], label=f'M={M}')

axes[1].set_title(r"Variação de $M$ (Mantendo $N, k$ fixos)")
axes[1].set_xlabel('X')
axes[1].legend()
axes[1].grid(alpha=0.5)

# Variando N, mantendo M e k fixos
for idx, N in enumerate(N_values):
    
    x_values = np.arange(max(0, k_base - (N - M_base)), min(k_base, M_base) + 1)
    pmf_values = stats.hypergeom.pmf(x_values, N, M_base, k_base)
    stem_container = axes[2].stem(x_values, pmf_values, basefmt="C0-", markerfmt='o', linefmt='-')
    plt.setp(stem_container.markerline, color=cores[idx])
    plt.setp(stem_container.stemlines, color=cores[idx])
    axes[2].plot(x_values, pmf_values, color=cores[idx], label=f'N={N}')

axes[2].set_title(r"Variação de $N$ (Mantendo $M, k$ fixos)")
axes[2].set_xlabel('X')
axes[2].legend()
axes[2].grid(alpha=0.5)

plt.show()
```
